## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,

## Solution 1

```python
#Loading the json data into a variable
dataset = pd.read_json('world_bank_projects.json')

#setting up the display to show all the columns
pd.set_option('display.max_columns',50)
dataset.head()
```

All the columns of the dataset are on display. On closer inspection, to determine the variables that would be useful to find the solution to first problem, Country name and project name columns stand out. Now the summary of the dataset could give some more useful insights.

```python
dataset.shape
```

```python
dataset.info()
```

Few of the columns in this dataset has missing values, but the choosen variables for the first task don't seem to have any missing values.

```python
# no. of unique countries
dataset['countryname'].nunique()
```

```python
# no. of unique projects
dataset['project_name'].nunique()
```

Since each entry in the dataset corresponds to a unique project and the unique no. of countries have also been identified. We can sort the dataframe countrywise and calculate the number of different projects associated with each country. On sorting it in descending order,we'll have the solution to the first task.

```python
dataset['countryname'].value_counts(sort=True,ascending=False).head(10)
```

## Solution 2

```python
# extracting the data in the 'mjtheme_namecode' column which appears to be a json object
data = dataset['mjtheme_namecode'][:]
data 
```

```python
# Inspecting an instance from this data
data[2]
```

```python
# converting a sample json object to a table to see the table and its format
json_normalize(data[2])
```

```python
# creating a table which contains all the normalized json objects in one dataframe
dfs = []
for x in range(0,500):
    d=json_normalize(data[x])
    dfs.append(d)
    
dataset1 = pd.concat(dfs,ignore_index=True)
dataset1.head(25)

    
```

```python
dataset1.info()
```

Now sorting the dataframe on the 'name' column to find the top 10 themes of the various projects undertaken

```python
dataset1.name.value_counts(sort=True,ascending=False).head(10)
```

## Solution 3


Now, It has been mentioned that some of the values in the  'mjtheme_namecode' dataframe under the name column are absent and the rows has been filled with a blank. We have to look for these rows with missing values and fill it with relevant names and store the result in a dataframe.

```python
# checking for how many unique codes are there
dataset1.code.nunique()
```

```python
# checking for how many unique project themes are there
dataset1.name.unique()
```

As it can be easily inferred, the 11 different codes are associated with 11 different project themes with a blank also being present.

```python
#checking for the no. of missing entries
miss_data = dataset1[dataset1['name']=='']
miss_data
```

By taking a look at the top 25 first instances of the 'mjtheme_namecode' dataframe, and comparing it with 'miss_data' dataframe which has missing values, the 'code' and relevant 'name' are consistent. Now in the next step,rows with missing values are going to be dropped and then all the duplicates are going to be dropped from the dataframe.

```python
dmjtheme = dataset1.copy()
```

```python
# dropping all the rows that have missing values
dmjtheme.drop(dmjtheme[dmjtheme['name']==''].index,inplace=True)
dmjtheme
```

```python
# dropping all the duplicates from the dataframe to see the association of each code to a project theme
dmjtheme.drop_duplicates(inplace=True)
dmjtheme = dmjtheme.reset_index().drop('index',axis=1)
dmjtheme
```

The above table shows the unique names of the themes with their associated code numbers. Now merging this template dataframe with the original to fill the missing values with relevant names.

```python
dataset1 = pd.merge(dataset1,dmjtheme,left_on='code',right_on='code',how='left')
dataset1
```

```python
dataset1 = dataset1.drop('name_x',axis=1)
dataset1
```

```python
dataset1 = dataset1.rename(columns={'name_y':'name'})
dataset1
```

Checking if any blank spaces are present 

```python
dataset1[dataset1.name==''].count()
```
